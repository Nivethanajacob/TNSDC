```python
string = input("Enter a string value: ")

char_count = {}

for ch in string:
    if ch in char_count:
        char_count[ch] += 1
    else:
        char_count[ch] = 1


for ch, count in char_count.items():
    print(f"{ch} = {count}", end=", ")
```

```python
def find_max(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

print(find_max(34, 12, 7))
```

```python
def exponent(base, exp):
    return base ** exp

b = int(input("Enter the base: "))
e = int(input("Enter the exponent: "))

print(exponent(b, e))
```

```python
def sum_of_cubes(n):
    total = 0
    for i in range(1, n):
        total += i**3
    return total

num = int(input("Enter a positive integer: "))
print(sum_of_cubes(num))
```
